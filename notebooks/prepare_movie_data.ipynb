## Custom `read_tsv()` Function

We define a custom `read_tsv()` function based on pandas `read_csv()` method to properly load tab separated (TSV) files.

The raw data files do not include header rows, so using `read_csv()` directly would result in unnamed columns. To address this, we specify the column names when loading the data.

````python
# Define function to read in tsv file with correct columns
def read_tsv(file_path, columns):
    """
    Read a tab-separated (TSV) file into a pandas DataFrame
    using a user-provided list of column names.

    Parameters
    -----------
        file_path : str
            Path to the TSV file.
        columns : list
            Column names to assign to the DataFrame.

    Returns
    --------
        pandas.DataFrame
            DataFrame containing the loaded TSV data
            with the specified column names.
    """
    return pd.read_csv(
        file_path,
        sep= "\t",
        header= None,
        names = columns
    )
````

````python
# Initialize list of movie, character, and plot columns
movie_metadata_cols = [
    "wiki_movie_id",
    "freebase_movie_id",
    "movie_name",
    "release_date",
    "bo_revenue",
    "runtime",
    "languages",
    "countries",
    "genres"
]

character_metadata_cols = [
    "wiki_movie_id",
    "freebase_movie_id",
    "movie_release_date",
    "character_name",
    "actor_dob",
    "actor_gender",
    "actor_height",
    "actor_ethnicity",
    "actor_name",
    "actor_age_at_release",
    "freebase_character_map",
    "freebase_character_id",
    "freebase_actor_id"
]

plot_cols = [
    "wiki_movie_id",
    "plot"
]
````

# Transform Data

### Custom `get_dict_values()` Function

We defined a custom `get_dict_values()` function to extract and format the values stored in specific text columns.

Some columns in the ``movies`` dataset such as `language`, `countries`, and `genres` contain dictionary data stored as strings. The function converts each string into a dictionary using `ast.literal_eval()`, then extracts the dictionary values and joins them into a single comma-separated string.

**Reference:**
- https://www.geeksforgeeks.org/python/ways-to-convert-string-to-dictionary/

````python
# Define function to get values from dictionary
import ast
def get_dict_values(col):
    """
        Convert a dictionary-like string into a comma-separated string of its values.
        
        Parameters
        ----------
            col : str 
                A dictionary-like string or missing value.

        Returns
        -------
            str
                A comma-separated string of dictionary values, or an empty string
                if the input is missing.
        """
    # Check that the column isn't missing values
    if pd.isna(col):
        return ""
    else:
        # Otherwise return the formated string
        dictionary = ast.literal_eval(col)
        return ", ".join(dictionary.values())
````

````python
# Call on movies
movies['languages'] = movies['languages'].apply(get_dict_values)
movies['countries'] = movies['countries'].apply(get_dict_values)
movies['genres'] = movies['genres'].apply(get_dict_values)
````

````python
# Remove "Language" from languages column in movies dataframe
movies['languages'] = movies['languages'].str.replace("Language", "")
````

## Creating a Cast Dataset

We create a smaller copy of the `characters` dataframe that keeps only the most relevant fields for the RAG pipeline: `wiki_movie_id`, `character_name`, and `actor_name`. We wanted to include only information needed to connet movies to their characters and actors. 

````python
# Create a copy of the character dataframe and subset
characters_small = characters[['wiki_movie_id', "character_name", "actor_name"]].copy()

# Basic cleaning (removing leading and lagging whitespaces)
for col in ['character_name', "actor_name"]:
    characters_small[col] = characters_small[col].str.strip()
````

Example:
``Katniss Everdeen: Jennifer Lawrence``

````python
# Replace missing data
characters_small['actor_name'] = characters_small['actor_name'].fillna("Unknown Actor")
characters_small['character_name'] = characters_small['character_name'].fillna("Unknown Character")

# Create character_actor column
characters_small['character_actor'] = characters_small['character_name'] + ": " + characters_small['actor_name']
````

## Aggregating Cast Information by Movie

At this stage, the character-actor data still contains multiple rows per movie, with each row representing a single character and actor pair. To prepare the data for retrieval, we aggregate this information so that each movie has only one row.

We group the data by `wiki_movie_id` and collect all `character_actor` entries for the same movie into a list using `apply(list)`. This creates a single record per movie while preserving all associated character-actor pairs.

Afterwards `reset_index()` converts the grouped result back into a pandas dataframe.

````python
# Aggregate to one row per movie
characters_agg = characters_small.groupby('wiki_movie_id')['character_actor'].apply(list).reset_index()

# View result
print(characters_agg.iloc[3]['character_actor'][:5])

# Print a couple rows
characters_agg.head()
````

## Reformatting Cast Entries

After aggregation, each movie contains a list of `character_actor` strings. However, some characters may appear more than once if multiple actors are associated with the same role.

To make this information cleaner and easier to read, we define a custom `reformat_character_actor()` function, it creates a dictionary in which each key is the character and the values are the actors who played that character. 

Overall this a more readable cast representation and creates cleaner text for cast-based chunks. This is the dataset we'll later use for merging. 

````python
# Create function to reformat character_actor
def reformat_character_actor(character_actor):
    """
    Reformat a list of character-actor strings into a structured multiline string.

    This function takes a list of strings in the format `character:actor`,
    groups actors by character name, and returns a cleaner text representation
    where each character appears once followed by its associated actor or actors.

    Parameters
    ----------
        character_actor : list
            List of character-actor strings.

    Returns
    --------
        str
            Multiline string where each line contains a character name
            followed by one or more associated actor names.
    """
    # Initialie dictionary to hold character-actor 
    character_dict = {}

    # Split each "character:actor" entry and group actors under each character
    for i in character_actor:
        parts = i.split(":")
        character = parts[0]
        actor = parts[1]

        # Populate dictionary keys
        if character not in character_dict:
            character_dict[character] = []
        
        # Add values to the dictionary
        character_dict[character].append(actor)

    # Initialize list to store clean strings
    clean_format = []

    # Rebuild the grouped data into a readable "character: actor1, actor2" format
    for character, actor in character_dict.items():
        clean_format.append(f"{character}: {', '.join(actor)}")

    # Return one multiline string per movie
    return "\n".join(clean_format)

# Apply formatting function to aggregated character-actor column
characters_agg['character_actor'] = characters_agg['character_actor'].apply(reformat_character_actor)
````

## Encoding Text Columns
Before continuing with preprocessing, we standardize text values across the dataset by applying a custom `encode_text()` function to all object-type columns.

This function checks whether a value is a string. If so, it encodes the text as UTF-8 and ignores any invalid characters before decoding it back into a clean string. We did this because we ran into some encoding errors and wanted to ensure that our encoding was standardized across all entries. 

**Reference:**
- https://www.geeksforgeeks.org/machine-learning/python-efficient-text-data-cleaning/

````python
# Encode
def encode_text(text):
    """
    Clean a text value by enforcing UTF-8 encoding.

    This function checks whether the input is a string. If so, it encodes
    the text in UTF-8 while ignoring invalid characters, then decodes it
    back into a cleaned string. Non-string values are returned unchanged.

    Parameters
    -----------
        text : str 
            Input string to clean.

    Returns
    -------
        str
            UTF-8 cleaned string if the input is text, otherwise
            the original value.
    """
    # Ensure input is a string
    if type(text) == str:
        # Encode and decode
        cleaned_text = text.encode("utf-8", "ignore").decode("utf-8")
        return cleaned_text
    else:
        return text
````

````python
# Iterate through columns
for col in combo_movies.columns:
    # Make sure dtype is object
    if combo_movies[col].dtype == "object":
        # Apply function
        combo_movies[col] = combo_movies[col].apply(encode_text)
````

## Plot Cleaning Function

We define a custom `clean_plot_text()` function to remove common formatting artifacts and markup from the raw movie plot summaries.

This function cleans raw movie plot text by:
- Decoding HTML entities
- Removing wiki templates and reference tags
- Stripping HTML and URLs
- Converting wiki links to readable text
- Normalizing whitespace

For a more detailed explanation of the regex patterns and cleaning steps, see `plot_cleaning_breakdown.md`.

````python
# Import regular expressions and html
import re
import html

# Define function for general plot cleaning
def clean_plot_text(text):
    """
    Clean raw movie plot text by decoding html, removing wiki templates,
    ref tags, and other unwanted text.

    This function is designed for plot summaries that contain wiki templates, 
    HTML tags, reference tags, URLs, wiki-style links, escaped characters, 
    and irregular whitespace. It applies a series of regex-based cleaning steps 
    to produce more readable and consistent text for downstream systems.

    Parameters
    ----------
        text : str
            Raw plot text

    Returns
    --------
        str or None
            Cleaned plot text, or None if the input is missing
            or becomes empty after cleaning.
    """
    # If plot is null
    if pd.isna(text):
        return None

    # Ensure plot is a string
    text = str(text)

    # Decode html entities
    text = html.unescape(text)

    # Remove common wiki templates ({{plot}}, {{tone}}, {{Expand section}}, {{long plot}})
    text = re.sub(r"\{\{[^{}]*\}\}", " ", text)

    # Remove ref tags like <ref>...</ref> and self-closing refs
    text = re.sub(r"<ref[^>]*>.*?</ref>", " ", text, flags=re.IGNORECASE | re.DOTALL)
    text = re.sub(r"<ref[^>]*/\s*>", " ", text, flags=re.IGNORECASE)

    # Remove other HTML tags
    text = re.sub(r"<[^>]+>", " ", text)

    # Remove URLs
    text = re.sub(r"http\S+|www\.\S+", " ", text)

    # Removes actors mentioned following a character
    text = re.sub(r"\(\[\[\s*[A-Z][a-z]+(?:[\s_]+[A-Z][a-z]+){0,3}(?:_+)?\s*,?", "", text) 
    
    # Convert wiki links
    text = re.sub(r"\[\[([^|\]]+)\|([^\]]+)\]\]", r"\2", text)
    text = re.sub(r"\[\[([^\]]+)\]\]", r"\1", text)

    # Remove leftover broken wiki brackets (just in case any where missed)
    text = text.replace("[[", " ").replace("]]", " ")

    # Fix common escaped characters
    text = text.replace("\\'", "'")
    
    # Normalize whitespace
    text = re.sub(r"\s+", " ", text).strip()
    
    # For empty strings
    if text == "":
        return None

    return text
````

### Handling Movies with the Same Title

While checking for duplicates in the movie dataset, we found that some movies share the same title. However these aren't cases of true duplicates, since some movies have remakes or reboots.

Because of this, we chose **not drop duplicate movie titles**. Instead, we will add a unique identifier such as: 
- **Wikipedia ID**
- **Movie Title**
- **Release Date**

to each chunk.

This should give the LLM enough information to differeniate between movies with identical names.
